# Keras implementation of YOLO v1

```python
import tensorflow as tf
import cv2
import numpy as np
from matplotlib import pyplot as plt

print(tf.__version__)
print(tf.test.is_gpu_available())
```

### 훈련 데이터셋

```python
# Get img list
with open('./custom_data/train.txt', 'r') as file:
    img_list = file.readline()
    print('img_list: ', img_list)

# Get img and convert BGR to RGB
# OpenCV -> BGR
# matplotlib -> RGB
train_img = cv2.imread(img_list[0])
train_img = cv2.cvtColor(train_img, cv2.COLOR_BGR2RGB)

# Get img shape
img_height, img_width, channel = train_img.shape
print('train_img shape: ', train_img.shape)


# Show img
plt.imshow(train_img)
plt.title('Train Img')
plt.show()
```
